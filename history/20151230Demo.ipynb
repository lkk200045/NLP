```python
def setNewsAry(soup):
    for news in soup.select('.rtddt'):   
        newsurl = news.select('a')[0]['href']
        if domain not in newsurl:
            newsurl = domain + newsurl
        main = news.select('h1')[0].text
        m = re.match('(.*)\((\d+)\)',main)
        if m:
            title = m.group(1)
            popularity = int(m.group(2))
        category = news.select('h2')[0].text 
        time = news.select('time')[0].text 
        url = newsurl
        dic = {'title': title, 'popularity':popularity,'category':category, 'time':time, 'url':url}
        newsary.append(dic)
```

```python
import requests
from bs4 import BeautifulSoup 
newsary = []

for i in range(1,51):
    pageurl = 'http://www.appledaily.com.tw/realtimenews/section/new/{}'
    newsurl = pageurl.format(i)
    res = requests.get(newsurl)
    soup = BeautifulSoup(res.text)
    setNewsAry(soup)

print len(newsary)
```

```python
# -*- coding: utf-8 -*- 
import sqlite3
db = sqlite3.connect('news.sqlite') 
cur = db.cursor()
cur.execute('''DROP TABLE IF EXISTS news_entry''')
cur.execute('''CREATE TABLE IF NOT EXISTS news_entry(
                id INTEGER PRIMARY KEY,
                category TEXT,
                popularity INTEGER,
                time VARCHAR(32),
                title VARCHAR(256),
                summary TEXT,
                url VARCHAR(1024)
                )''')
cur.close()
```

```python
db = sqlite3.connect('news.sqlite') 
cur = db.cursor()

for news in newsary:
    col = ', ' .join(news.keys())
    val = ','.join('?'* (len(news.keys())))
    sql = 'insert into news_entry(%s) values(%s)'%(col, val)
    cur.execute(sql,news.values())
    db.commit()
          
cur.close()
db.close()
```
